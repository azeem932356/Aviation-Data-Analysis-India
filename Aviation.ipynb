```python
#Convert 'Date' column to datetime
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')

#Drop rows with invalid/missing dates
df = df.dropna(subset=['Date'])

#Filter for 2020-07-07 to 2021-03-31
mask = (df['Date'] >= '2020-07-07') & (df['Date'] <= '2021-03-31')
df_filtered = df.loc[mask].copy()

#Sort for smooth plotting
df_filtered = df_filtered.sort_values('Date')
```

### Domestic Airport Footfalls (Jul 2020 – Mar 2021) Before COVID-19 --Airport Footfalls

```python
# =========================
# Chart 1: Airport Footfalls
# =========================
plt.figure(figsize=(12, 5))
plt.plot(df_filtered['Date'], df_filtered['Domestic (Airport Footfalls)'], marker='o', color='steelblue')
plt.title('Domestic Airport Footfalls (Jul 2020 – Mar 2021) Before COVID-19')
plt.xlabel('Monthly Wise Data')
plt.ylabel('Total number of passengers(arrivals + departures)')
plt.grid(True)
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

### Domestic Aircraft Movements (Jul 2020 – Mar 2021) Before COVID-19 --Aircraft Movements

```python
# =========================
# Chart 2: Aircraft Movements
# =========================
plt.figure(figsize=(12, 5))
plt.plot(df_filtered['Date'], df_filtered['Domestic (Aircraft Movements)'], marker='s', color='darkred')
plt.title('Domestic Aircraft Movements (Jul 2020 – Mar 2021) Before COVID-19')
plt.xlabel('Monthly Wise Data')
plt.ylabel('Total number of takeoffs and landings')
plt.grid(True)
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

### International Passenger Movement (Arriving vs Departing)

```python
plt.figure(figsize=(12, 6))

# Plot Arrivals
plt.plot(df['Date'], df['International (Arriving Pax)'], label='Arriving Pax', color='blue')

# Plot Departures
plt.plot(df['Date'], df['International (Departing Pax)'], label='Departing Pax', color='orange')

plt.title('International Passenger Movement (Arriving vs Departing)', fontsize=14)
plt.xlabel('Date')
plt.ylabel('Passenger Count')
plt.grid(True)
plt.xticks(rotation=0)
plt.legend()
plt.tight_layout()
plt.show()
```

### Monthly Domestic Arrival vs Departure Flights (Jul 2020 – Jun 2025)

```python
# Filter the date range
start_date = '2020-07-07'
end_date = '2025-06-17'
df_filtered = df[(df['Date'] >= start_date) & (df['Date'] <= end_date)].copy()

# Set Date as index
df_filtered.set_index('Date', inplace=True)

# Resample monthly and take the sum
monthly_df = df_filtered.resample('M').sum()

# Plotting
plt.figure(figsize=(14, 6))
plt.plot(monthly_df.index, monthly_df['Domestic (Arrival Flights)'],
         label='Arrival Flights', color='blue', linewidth=2, marker='o')

plt.plot(monthly_df.index, monthly_df['Domestic (Departure Flights)'],
         label='Departure Flights', color='orange', linewidth=2, marker='o')

plt.title('Monthly Domestic Arrival vs Departure Flights (Jul 2020 – Jun 2025)')
plt.xlabel('Month')
plt.ylabel('Number of Flights')
plt.legend()
plt.grid(True)
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

### Domestic vs International Aircraft Movements Over Time

```python
# Filter only the necessary columns
df_filtered = df[['Date', 'Domestic (Aircraft Movements)', 'International (Aircraft Movements)']]

# Sort by Date
df_filtered = df_filtered.sort_values('Date')

# Plot
plt.figure(figsize=(14, 6))
plt.plot(df_filtered['Date'], df_filtered['Domestic (Aircraft Movements)'], label='Domestic Aircraft Movements', color='blue')
plt.plot(df_filtered['Date'], df_filtered['International (Aircraft Movements)'], label='International Aircraft Movements', color='orange')

plt.title('Domestic vs International Aircraft Movements Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Aircraft Movements')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.xticks(rotation=0)
plt.show()
```

### Domestic Arrival vs Departure Flights Over Time

### International Arrival vs Departure Flights Over Time

```python
# Plot International
plt.figure(figsize=(14, 6))
plt.plot(df_filtered['Date'], df_filtered['International (Arrival Flights)'], label='International Arrival Flights', color='orange')
plt.plot(df_filtered['Date'], df_filtered['International (Departure Flights)'], label='International Departure Flights', color='blue')

plt.title('International Arrival vs Departure Flights Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Flights')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.xticks(rotation=0)
plt.show()
```

### Airport Footfalls vs Total Passenger Numbers (Domestic)

```python
# Sort by Date
df = df.sort_values('Date')

# Create Total Passenger column
df['Total Pax'] = df['Domestic (Arriving Pax)'] + df['Domestic (Departing Pax)']

# Plotting
plt.figure(figsize=(14, 6))
plt.plot(df['Date'], df['Domestic (Airport Footfalls)'], label='Airport Footfalls', color='blue')
plt.plot(df['Date'], df['Total Pax'], label='Passenger Numbers (Arriving + Departing)', color='green')

plt.title('Airport Footfalls vs Total Passenger Numbers (Domestic)')
plt.xlabel('Date')
plt.ylabel('Number of Passengers')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.xticks(rotation=45)
plt.show()
```

## Year Wise Analysis

```python
df.head(20)
```

```python
# Grouping by Year
yearly_data = df.groupby('Year')[['PaxToIndia', 'PaxFromIndia']].sum().reset_index()

# Convert to Millions
yearly_data['PaxToIndia'] = yearly_data['PaxToIndia'] / 1_000_000
yearly_data['PaxFromIndia'] = yearly_data['PaxFromIndia'] / 1_000_000

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(yearly_data['Year'], yearly_data['PaxToIndia'], label='To India', linewidth=3)
plt.plot(yearly_data['Year'], yearly_data['PaxFromIndia'], label='From India', linewidth=3)

plt.title('Year-wise Total Passengers (in Millions)', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Passengers (Millions)', fontsize=12)
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
